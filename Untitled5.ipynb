```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_theme(style="whitegrid") 
```

```python
df=pd.read_csv("/content/train_data.csv")
```

```python
lab=list(df['airline'].unique())
values=list(df['airline'].value_counts())
plt.figure(figsize=(5,5))
plt.pie(values, labels=lab, shadow=True)
plt.show()
```

```python
df['airline'].value_counts()
```

```python
fig, ax=plt.subplots(3,2, figsize=(16,16))
sns.countplot(ax=ax[0, 0], data=df, x='source_city')
sns.countplot(ax=ax[0, 1], data=df, x='destination_city')
sns.countplot(ax=ax[1, 0], data=df, x='departure_time')
sns.countplot(ax=ax[1, 1], data=df, x='arrival_time')
sns.countplot(ax=ax[2, 0], data=df, x='stops')
sns.countplot(ax=ax[2, 1], data=df, x='class');
```

```python
for i in range(len(df)):
  if df['source_city'][i]==df['destination_city'][i]:
    print('Error destination or location to get a flight!')
```

```python
df.drop(['id'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
df['flight'] = LE.fit_transform(df['flight'])
df['airline'] = LE.fit_transform(df['airline'])
df['source_city']=LE.fit_transform(df['source_city'])
df['departure_time']=LE.fit_transform(df['departure_time'])
df['stops']=LE.fit_transform(df['stops'])
df['arrival_time']=LE.fit_transform(df['arrival_time'])
df['destination_city']=LE.fit_transform(df['destination_city'])
df['class']=LE.fit_transform(df['class'])
df.head()
```

```python
df
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
strat_split=StratifiedShuffleSplit(n_splits=1, test_size=0.1, random_state=42)
for tr_set, ts_set in strat_split.split(df, df['class']):
  train_set_split=df.loc[tr_set]
  test_set_split=df.loc[ts_set]
```

```python
sns.countplot(data=train_set_split, x='class')
plt.show()
```

```python
sns.countplot(data=df, x='class')
plt.show()
```

```python

fig, ax = plt.subplots(2,3, figsize=(16, 14))

sns.countplot(ax=ax[0,0], data=train_set_split, x='class')
sns.countplot(ax=ax[0, 1], data=test_set_split, x='class')
sns.countplot(ax=ax[0,2], data=df, x='class')
```

```python
X_train=train_set_split.drop(['price'], axis=1)
y=train_set_split['price']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
num_pipeline=Pipeline([
    ('standart_scaler', StandardScaler())
])
```

```python
X_prepared=num_pipeline.fit_transform(X_train)
```

```python
X_prepared
```

```python
X_train.head()
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.linear_model import LinearRegression


LR_model = LinearRegression()

LR_model.fit(X_prepared, y)
X_test = test_set_split.drop('price', axis=1)
y_test = test_set_split['price'].copy()
X_test_prepared = num_pipeline.transform(X_test)
y_predicted = LR_model.predict(X_test_prepared)

mae = mean_absolute_error(y_test, y_predicted)
mse = mean_squared_error(y_test, y_predicted)

print("MAE=", np.around(mae))
print("RMSE=", np.around(np.sqrt(mse)))
```
